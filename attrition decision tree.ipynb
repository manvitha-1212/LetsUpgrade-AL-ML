```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
```

```python
data=pd.read_csv("general_data.csv")
```

```python
data.head()
```

```python
data['Gender'] = data['Gender'].map({'Male':1, 'Female': 0})
```

```python
data.head()
```

```python
data['Attrition'] = data['Attrition'].map({'Yes':1, 'No': 0})
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data["NumCompaniesWorked"].mean()
```

```python
data["TotalWorkingYears"].mean()
```

```python
new1=np.where(data["NumCompaniesWorked"].isnull(),2,data["NumCompaniesWorked"])
```

```python
data["NumCompaniesWorked"]=new1
```

```python
new2=np.where(data["TotalWorkingYears"].isnull(),11,data["TotalWorkingYears"])
```

```python
data["TotalWorkingYears"]=new2
```

```python
data.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data.columns
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
data["BusinessTravel"]=label_encoder.fit_transform(data["BusinessTravel"])
```

```python
data["Department"]=label_encoder.fit_transform(data["Department"])
```

```python
data["EducationField"]=label_encoder.fit_transform(data["EducationField"])
```

```python
data.head()
```

```python
data["JobRole"]=label_encoder.fit_transform(data["JobRole"])
```

```python
data["MaritalStatus"]=label_encoder.fit_transform(data["MaritalStatus"])
```

```python
data["Over18"]=label_encoder.fit_transform(data["Over18"])
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
features=['Age', 'BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'EmployeeCount', 'EmployeeID', 'Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'Over18', 'PercentSalaryHike', 'StandardHours',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
rf_model.fit(X=data[features],y=data["Attrition"])
```

```python
print("Accuracy")
print(rf_model.oob_score_);
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp);
```

We can choose Dependent variable to be - "Age","MonthlyIncome","TotalWorkingYears" as their values are comparitively higher.

```python
treemodel=tree.DecisionTreeClassifier(max_depth=4)
```

```python
data1=pd.DataFrame([data["TotalWorkingYears"],data["Age"],data["MonthlyIncome"]]).T
```

```python
treemodel.fit(X=data1,y=data["Attrition"])
```

```python
with open("decision_tree_2.dot",'w')as f:
    f=tree.export_graphviz(treemodel,feature_names=["TotalWorkingYears","Age","MonthlyIncome"],out_file=f);
```

```python
treemodel.score(X=data1,y=data["Attrition"])
```
